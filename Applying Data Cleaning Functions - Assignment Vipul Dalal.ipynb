```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('matches.csv')
df.head(2)
```

# Question 1.) Find out the number of venues

```python
df['venue'].nunique()
```

# Question 2.) Find out the number of times each team emerged as the winner.

```python
df.groupby('winner')[['id']].count().rename(columns = {'id' : 'No. of Wins'})
```

# Question 3.) Find out the times the toss winner chose bat and the number of times the winner chose field.

```python
df.groupby('toss_decision')[['id']].count().rename(columns = {'id' : 'Total No. of Decision'})
```

# Question 4.) Replace the toss_decision value with bowl_first where toss_decision is field.

```python
df[['toss_decision']].where(df['toss_decision'] != 'field', 'bowl_first')
```

# Question 5.) Sort the dataset in descending order of win by runs and print the column win_by_runs.

```python
df[['win_by_runs']].sort_values(by = 'win_by_runs', ascending = False)
```

# Question 6.) Find the date where the winner is Royal Challengers Bangalore with runs more than 30.

```python
r = df.query(" winner == 'Royal Challengers Bangalore' and win_by_runs > 30")
r[['date', 'winner', 'win_by_runs']]
```

# Question 7.) Find the max win by runs and win by wickets using apply function.

```python
df[['win_by_runs','win_by_wickets']].apply(np.max)
```

```python
import numpy as np
import pandas as pd
```

## Creating the Sample Data Frame

```python
student_data = pd.DataFrame({'school_code': ['s001','s002','s003','s001','s002','s004'],
                             'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
                             'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton',
                                      'Gino Mcneill', 'David Parkes'],
                             'date_Of_Birth ':['15/05/2002','17/05/2002','16/02/1999','25/09/1998',
                                               '11/05/2002','15/09/1997'],
                             'age': [12, 12, 13, 13, 14, 12],
                             'height': [173, 192, 186, 167, 151, 159],
                             'weight': [35, 32, 33, 30, 31, 32],
                             'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']},
                            index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
student_data
```

# Question 1.) Write a Pandas program to split the following dataframe by school code and get mean, min, and max value of age for each school.

```python
result_1 = student_data.groupby('school_code').agg({'age' : ['mean', 'min', 'max']})
result_1
```

# Question 2.) Write a Pandas program to split the following dataframe by school code and get median, variance, and standard deviation of age for each school.

```python
result_2 = student_data.groupby('school_code').agg({'age' : ['median', 'std', 'var']})
result_2
```

# Question 3.) Write a Pandas program to split the following dataframe into groups by school code and get mean, min, and max value of age with customized column name for each school.

```python
result_3 = student_data.groupby('school_code').agg(Age_Mean = ('age','mean'),
                                                   Age_Max = ('age','min'),
                                                   Age_Min = ('age','max'))
result_3
```
